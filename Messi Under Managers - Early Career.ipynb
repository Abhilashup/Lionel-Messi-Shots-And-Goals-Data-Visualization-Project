# INTRODUCTION

- The data for this part of the notebook was gathered through StatsBomb API and the data is available for free on their GitHub. They were gathered, seperated into different managers in an Excel Sheet and later used here for the analysis.

- The managers in this notebook are between 2004/05 season till the recent 2013/14 season.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from mplsoccer import VerticalPitch, Pitch #library used for football visualizations
from mplsoccer.utils import FontManager
from highlight_text import htext, fig_text, HighlightText
```

# FRANK RIJKAARD

```python
messi_frank_shots = pd.read_csv('./Data/messi_rijkaard.csv')
messi_frank_shots.head()
```

```python
#Filtering for goals to be a seperate dataframe and is done for every manager
messi_frank_goals = messi_frank_shots[messi_frank_shots['shot_outcome']=='Goal']
messi_frank_goals.head()
```

```python
#Calculating the total Expected Goals(xG) and Goals scored. This too is done for every manager.
total_xg_shots_one = np.round(messi_frank_shots['xG'].sum(), 2)
total_goals_one = len(messi_frank_goals)

total_xg_shots_one, total_goals_one
```

```python
# Using Eucledian distance to calculate the average distance of goals.
avg_x = messi_frank_goals['x'].mean()
avg_y = messi_frank_goals['y'].mean()
avg_distance_from_goal_one = np.sqrt(np.sum(np.square(np.array([120, 40])-
                                                    np.array([avg_x, avg_y]))))

avg_distance_from_goal_one = np.round(avg_distance_from_goal_one, 2)
avg_distance_from_goal_one
```

```python
# Using Eucledian distance to calculate the average distance of shots
avg_x = messi_frank_shots['x'].mean()
avg_y = messi_frank_shots['y'].mean()
avg_distance_shot_one = np.sqrt(np.sum(np.square(np.array([120, 40])-
                                                    np.array([avg_x, avg_y]))))

avg_distance_shot_one = np.round(avg_distance_shot_one, 2)
avg_distance_shot_one
```
